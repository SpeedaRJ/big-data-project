```python
import pandas as pd
import dask.dataframe as dd
```

```python
tickets = dd.read_parquet('../data/parking_tickets/parquet/full_data_cleaned.parquet')
weather = pd.read_csv('../data/additional_data/weather/weather_NYC_2013_2024_processed.csv')
```

```python
display(tickets.head()), display(weather)
tickets["Issue Date"].dtypes, weather["datetime"].dtypes
```

```python
tickets["Issue Date"] = dd.to_datetime(tickets["Issue Date"], unit="ms").dt.floor("D")
weather["datetime"] = pd.to_datetime(weather["datetime"]).dt.floor("D")
```

```python
display(tickets.head()), display(weather.head())
```

```python
merged = tickets.merge(weather, how="left", left_on="Issue Date", right_on="datetime")
merged
```
